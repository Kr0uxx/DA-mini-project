```python
X_train[numeric_cols] = X_train[numeric_cols].apply(pd.to_numeric, errors='coerce')
X_train[categorical_cols] = X_train[categorical_cols].astype(str)
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
    ])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
# логистическая регрессия
linear = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

# Запуск кросс-валидации с 5 фолдами и оценкой точности
# scores1 = cross_val_score(linear, X_train, y_train, cv=5, scoring='roc_auc')

param_grid = {
    'classifier__C': [0.01, 0.1, 1, 10],
    'classifier__penalty': ['l2'],
    'classifier__solver': ['lbfgs']
}
grid_search = GridSearchCV(
    estimator=linear,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

logistic_model = grid_search.fit(X_train, y_train)
```
