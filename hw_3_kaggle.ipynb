```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
X_origin = data.iloc[:, :-1]
X_origin.head()
```

```python
X = data.drop(columns=['Churn']); y = data.Churn
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.8,
                                                    random_state=42,
                                                    stratify=y)
X_train.head()
```

```python
data[numeric_cols]
```

```python
X_train[numeric_cols] = X_train[numeric_cols].apply(pd.to_numeric, errors='coerce')
X_train[categorical_cols] = X_train[categorical_cols].astype(str)
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
    ])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
#  
linear = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

#  -  5    
# scores1 = cross_val_score(linear, X_train, y_train, cv=5, scoring='roc_auc')

param_grid = {
    'classifier__C': [0.01, 0.1, 1, 10],
    'classifier__penalty': ['l2'],
    'classifier__solver': ['lbfgs']
}
grid_search = GridSearchCV(
    estimator=linear,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

logistic_model = grid_search.fit(X_train, y_train)
```

```python

X_train[numeric_cols] = X_train[numeric_cols].replace(' ', np.nan)
X_test[numeric_cols] = X_test[numeric_cols].replace(' ', np.nan)
```

```python
y_proba = logistic_model.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, y_proba)
```
